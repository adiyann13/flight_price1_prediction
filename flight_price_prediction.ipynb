```python
### cleaning each column
# 1 Date_of_Journey
# this is a date time data type but its present in object type so converting it to proper type
import datetime as dt
flight_datas['Date_of_Journey_Month'] = pd.to_datetime(flight_datas['Date_of_Journey']).dt.month
```

```python
flight_datas['Date_of_Journey_Day'] = pd.to_datetime(flight_datas['Date_of_Journey']).dt.day
```

```python
flight_datas['Date_of_Journey_year'] = pd.to_datetime(flight_datas['Date_of_Journey']).dt.year
```

```python
#since we have exte=racted the data from date of journey column we can drop it
# also we can drop the year column since its juts a 2019 year data
flight_datas.drop(columns=['Date_of_Journey', 'Date_of_Journey_year'],axis=1,inplace=True)
```

```python
flight_datas
```

```python
###2 Dep_Time
flight_datas['Dep_time_hours'] =pd.to_datetime(flight_datas['Dep_Time']).dt.hour
```

```python
flight_datas['Dept_time_mins'] = pd.to_datetime(flight_datas['Dep_Time']).dt.minute
```

```python
flight_datas
```

```python
## we can now drop the deptime column since we have extracted the data from time
flight_datas.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
flight_datas
```

```python
#### 3 Arrival_Time
flight_datas['Arrival_Time_hours'] = pd.to_datetime(flight_datas['Arrival_Time']).dt.hour
flight_datas['Arrival_Time_mins'] = pd.to_datetime(flight_datas['Arrival_Time']).dt.minute
```

```python
flight_datas
```

```python
## dropping the arrival time column
flight_datas.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
flight_datas
```

```python
### 4 Duration
duration = list(flight_datas['Duration'])
duration
for i in range(len(duration)):
    if len(duration[i].split())!=2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip()+" 0m"
        else:
             duration[i] = "0h "+duration[i]
             print(duration[i])
            
            
            
        
```

```python
duration_hours=[]
duration_mintutes = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    duration_mintutes.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
flight_datas["duration_hour"]=duration_hours
flight_datas["duration_mintutes"]=duration_mintutes
```

```python
flight_datas
```

```python
##dropping the duration column off
flight_datas.drop(['Duration'],axis=1,inplace=True)
```

```python
flight_datas
```

```python
##dropping the route column since we have all the information regarding arrival destination
flight_datas.drop(['Route'],axis=1,inplace=True)
```

```python
## since ther is no info for 78% of the passengers wqe can drop this column
flight_datas.drop(['Additional_Info'],axis=1,inplace=True)
```

```python
####6  encoding the Airline                   
flight_datas['Airline'].value_counts()
```

```python
Airlines = flight_datas['Airline']
Airlines
```

```python
###7  Source                   
Sources = flight_datas['Source']
Sources = pd.get_dummies(Sources,drop_first=True,dtype='int64')
```

```python
Sources
```

```python
###8 Destination
flight_datas['Destination'].value_counts()
Destinations = flight_datas['Destination']
Destinations
Destinations=pd.get_dummies(Destinations,drop_first=True,dtype='int64')
```

```python
Destinations.dtypes
```

```python
#### 9 Total_Stops              
flight_datas['Total_Stops'].value_counts()
flight_datas['Total_Stops'].replace({"1 stop":1,"non-stop":0,"2 stops":2,"3 stops":3,"4 stops":4},inplace=True)
```

```python
flight_datas=pd.concat([flight_datas,Airlines,Sources,Destinations],axis=1,)
```

```python
### model building
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flight_data_indep,flight_data_dep,random_state=101)
y_train
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_mod = RandomForestRegressor().fit(x_train,y_train)
random_forest_mod
```

```python
random_forest_mod.feature_importances_
```

```python
imp_feats=pd.DataFrame({"imp_feat_df":random_forest_mod.feature_importances_,"columns":flight_data_indep.columns})
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=imp_feats,y=imp_feats['columns'],x=imp_feats['imp_feat_df'])
plt.show()
```

```python
###predict
price_train_predict = random_forest_mod.predict(x_train)
price_test_predict = random_forest_mod.predict(x_test)
```

```python
##evaluate
from sklearn.metrics import r2_score
r2_score_train = r2_score(y_train ,price_train_predict )
r2_score_train
```

```python
r2_score_test = r2_score(y_test ,price_test_predict )
r2_score_test
```

```python
### linear model
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression().fit(x_train,y_train)
lr_model
```

```python
### predict
lr_model_train_Predict = lr_model.predict(x_train)
lr_model_test_Predict = lr_model.predict(x_test)
```

```python
#evaluate
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae_test=mean_absolute_error(y_test,lr_model_test_Predict)
mae_test
```

```python
mse_test = mean_squared_error(y_test,lr_model_test_Predict)
mse_test
```
